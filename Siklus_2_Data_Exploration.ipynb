# Data Exploration

## Summary Descriptive Statistics

Summary descriptive statistics adalah rangkuman properti dari sebuah data

Properti tersebut adalah berupa:
- frekuensi,
- modus,
- mean,
- range, dan
- standar deviasi (deviation standard).

Tampilkan informasi data menggunakan syntax berikut:

image.png

```python
# Tulis di sini
df.info()
```

Berdasarkan informasi yang ditampilkan, data memiliki:
- 67 baris
- 5 kolom:
    - Sex dengan tipe data object/str
    - Weight (Sep) dengan tipe data int
    - Weight (Apr) dengan tipe data int
    - BMI (Sep) dengan tipe data float
    - BMI (Apr) dengan tipe data float

---

Pada Python, kita tidak perlu membuat program untuk frekuensi, mean, standar deviasi, range, dan kuartil.

Kita hanya perlu memanggil method `describe()`.


Method `describe()` secara otomatis menghitung:
- frekuensi (count)
- mean
- median (kuartil 50%)
- standar deviasi
- range (nilai min dan max)
- kuartil

Kolom-kolom yang **datanya numerik**.

image.png

```python
# Tulis di sini
df.describe()
```

Coba simpulkan hasil `describe()` di atas!



---

Selain membuat deskripsi kolom numerik, kita juga bisa membuat deskripsi kolom kategorik dengan menambahkan argumen `include='all'` pada `describe()`.



image.png

```python
# Tulis di sini
df.describe(include='object')
```

Coba simpulkan hasil `describe(include='object')` di atas!

## Data Visualization

Data Visualization (visualisasi data) adalah tampilan berupa grafis atau visual dari informasi dan data.

Dalam kata lain, data visualization mengubah kumpulan data menjadi hal lebih sederhana untuk ditampilkan.

Dengan menggunakan elemen visual tersebut, pembaca akan lebih mudah memahami tren, pola, dan outliers dalam suatu data.

Pada Python, terdapat dua library yang umum digunakan untuk membuat Data Visualization, yakni:
1. [Matplotlib](https://matplotlib.org/stable/index.html)
2. [Seaborn](https://seaborn.pydata.org/)

### Visualisasi Data Menggunakan Library Matplotlib

image.png

```python
# Tulis di sini
import matplotlib.pyplot as plt
```

#### 1. Bar Graph atau Bar Chart

Biasa digunakan untuk membandingkan kategori.

image.png

```python
# Tulis di sini
jumlah_pria = df['Sex'].value_counts()['M']
jumlah_wanita = df['Sex'].value_counts()['F']

plt.bar(['Pria', 'Wanita'], {jumlah_pria, jumlah_wanita})
plt.show()
```

Kita dapat mengubah ukuran dan menambah keterangan lain dalam plot, seperti:
- Judul plot
- Nama sumbu x
- Nama sumbu y
- Warna Bar Chart

image.png

```python
# Tulis di sini
plt.figure(figsize=(10, 3))
plt.bar(['Pria', 'Wanita'], [jumlah_pria, jumlah_wanita], color=['orange', 'pink'])
plt.title('Perbandingan jumlah Pria dan Wanita')
plt.xlabel('kategori')
plt.ylabel('Jumlah')
plt.show()
```

Contoh bar graph/chart lainnya:



image.png

```python
# Tulis di sini
rata2berat_pria = df[df['Sex']=='M'][' "Weight (Sep)"'].mean()
rata2berat_wanita = df[df['Sex']=='F'][' "Weight (Sep)"'].mean()
weights = [rata2berat_pria, rata2berat_wanita]
labels = ['Pria', 'Wanita']

plt.figure(figsize=(7, 5))
plt.bar(labels, weights)
plt.title('Perbandingan rata-rata berat badan pria dan wanita bulan September')
plt.show()
```

#### 2. Histogram

Biasa digunakan untuk menampilkan distribusi frekuensi.

Jalankan program berikut kemudian simpulkan plot-nya!

image.png

```python
# Tulis di sini
plt.hist(df[' "Weight (Sep)"'])
plt.title('Histogram dari berat badan bulan April')
plt.ylabel('Frekuensi')
plt.xlabel('Berat')
plt.show()
```

Cara lain dalam membuat histogram:

Jalankan program lalu simpulkan plot-nya!

image.png

```python
# Tulis di sini
df[' "Weight (Apr)"'].plot(kind='hist', figsize=(7, 5))
plt.title('Histogram dari berat badan bulan April')
plt.ylabel('Frekuensi')
plt.xlabel('Berat')
plt.show()
```

Apa perbedaan antara bar graph dan histogram?

#### 3. Line Chart

Biasa digunakan untuk menunjukkan tren pada data atau membandingkan data.

Contoh line chart:

image.png

```python
# Tulis di sini
bmi_april = df[' "BMI (Sep)"']

plt.plot(bmi_april)
plt.title('Perbandingan BMI bulan April dan September')
plt.ylabel('Peserta')
plt.xlabel('BMI')
plt.show()
```

Contoh line chart untuk membandingkan data:



image.png

```python
# Tulis di sini
weight_september = df[' "Weight (Sep)"']
bmi_september = df[' "BMI (Sep)"']

plt.plot(weight_september) #chart 1
plt.plot(bmi_september) #chart 2
plt.title('Perbandingan Berat dan BMI bulan September', fontsize=10)
plt.ylabel('Peserta', fontsize = 10)
plt.xlabel('Nilai', fontsize = 10)
plt.show()
```

Kita dapat menambahkan legenda dalam plot.



image.png

```python
# Tulis di sini
weight_september = df[' "Weight (Sep)"']
bmi_september = df[' "BMI (Sep)"']

plt.plot(weight_september, label='Berat bulan September') #chart 1
plt.plot(bmi_september, label='BMI bulan September') #chart 2
plt.title('Perbandingan Berat dan BMI bulan September', fontsize=10)
plt.ylabel('Peserta', fontsize = 10)
plt.xlabel('Nilai', fontsize = 10)
plt.legend()
plt.show()
```

Kita juga dapat mengubah bentuk dan warna garis dalam plot:

image.png

```python
# Tulis di sini
weight_september = df[' "Weight (Sep)"']
bmi_september = df[' "BMI (Sep)"']

plt.plot(weight_september, 'o--', label='Berat bulan September') #chart 1
plt.plot(bmi_september, 'r--', label='BMI bulan September') #chart 2
plt.title('Perbandingan Berat dan BMI bulan September', fontsize=10)
plt.ylabel('Peserta', fontsize = 10)
plt.xlabel('Nilai', fontsize = 10)
plt.legend()
plt.show()
```

Cara lain dalam membuat line chart:


image.png

```python
# Tulis di sini
df[[' "BMI (Apr)"', ' "BMI (Sep)"']].plot(kind='line', figsize=(10, 5))
plt.title('Perbandingan Berat dan BMI bulan September', fontsize=10)
plt.ylabel('Peserta', fontsize = 10)
plt.xlabel('BMI', fontsize = 10)
plt.show()
```

#### 4. Scatter plot

Menyajikan hubungan antara dua variabel dalam kumpulan data.

Contohnya menampilkan hubungan antara Berat dan BMI bulan April:

image.png

```python
# Tulis di sini
berat_september = df[' "Weight (Apr)"']
bmi_september = df[' "BMI (Apr)"']

plt.scatter(berat_september, bmi_september)
plt.title('Hubungan Berat dan BMI bulan April')
plt.xlabel('Berat', fontsize = 10)
plt.ylabel('BMI', fontsize = 10)
plt.show()
```

### Visualisasi Data Menggunakan Library Seaborn

```python
import seaborn as sns


# Data
categories = ['NS', 'EW', 'City', 'State']
counts = [128, 128, 128, 128]
unique_counts = [1, 1, 120, 47]
top_values = ['N', 'W', 'Springfield', 'CA']
top_freqs = [128, 128, 4, 12]

# Line Chart with Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x=categories, y=unique_counts, marker='o')
plt.xlabel('Categories')
plt.ylabel('Unique Count')
plt.title('Line Chart of Unique Counts (Seaborn)')
plt.grid(True)
plt.show()

# Scatter Plot with Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=categories, y=counts, color='red', label='Count')
sns.scatterplot(x=categories, y=unique_counts, color='blue', label='Unique')
sns.scatterplot(x=categories, y=top_freqs, color='green', label='Top Frequency')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Scatter Plot of Data Categories (Seaborn)')
plt.legend()
plt.show()
```

```python
import seaborn as sns


# Data
categories = ['NS', 'EW', 'City', 'State']
counts = [128, 128, 128, 128]
unique_counts = [1, 1, 120, 47]
top_values = ['N', 'W', 'Springfield', 'CA']
top_freqs = [128, 128, 4, 12]

# Histogram with Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(counts, bins=len(categories), kde=True)
plt.xlabel('Counts')
plt.ylabel('Frequency')
plt.title('Histogram of Counts (Seaborn)')
plt.show()

# Box Plot with Seaborn
data = {'Categories': categories * 3,
        'Values': counts + unique_counts + top_freqs,
        'Type': ['Count'] * len(categories) + ['Unique'] * len(categories) + ['Top Frequency'] * len(categories)}
plt.figure(figsize=(10, 6))
sns.boxplot(x='Categories', y='Values', hue='Type', data=data)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot of Data Categories (Seaborn)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
categories = ['NS', 'EW', 'City', 'State']
counts = [128, 128, 128, 128]
unique_counts = [1, 1, 120, 47]
top_values = ['N', 'W', 'Springfield', 'CA']
top_freqs = [128, 128, 4, 12]

# Cat Plot with Seaborn
plt.figure(figsize=(10, 6))
sns.catplot(data=[counts, unique_counts, top_freqs], kind="point", aspect=2.5)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Cat Plot of Data Categories (Seaborn)')
plt.xticks(ticks=range(len(categories)), labels=categories)
plt.show()

# Heatmap with Seaborn
data = [counts, unique_counts, top_freqs]
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, fmt='d', cmap='coolwarm', xticklabels=categories)
plt.xlabel('Categories')
plt.ylabel('Metrics')
plt.title('Heatmap of Data Categories (Seaborn)')
plt.show()
```

Histogram:

Histogram menunjukkan distribusi frekuensi dari jumlah data dalam setiap kategori.
Kategori "NS" dan "EW" memiliki hanya satu nilai unik dengan jumlah data yang sama, sementara "City" memiliki sebaran yang lebih bervariasi dengan banyak nilai unik.
Histogram untuk kategori "State" menunjukkan variasi yang lebih besar dalam distribusi negara bagian.
Box Plot:

Box plot memberikan gambaran tentang distribusi data secara lebih rinci, termasuk nilai median, kuartil, dan adanya pencilan (outlier).
Kategori "NS" dan "EW" memiliki nilai-nilai yang seragam dengan sedikit variasi, sementara "City" memiliki sejumlah besar variasi dan beberapa pencilan.
Distribusi data untuk kategori "State" menunjukkan variasi yang signifikan, dengan beberapa negara bagian memiliki frekuensi yang jauh lebih tinggi dari yang lain.
Cat Plot:

Cat plot menunjukkan tren perubahan dalam nilai-nilai kategori dari satu kategori ke kategori lain.
Dari grafik cat plot, kita dapat melihat perbedaan dalam pola antara jumlah total data, jumlah nilai unik, dan frekuensi nilai teratas untuk setiap kategori.
Heatmap:

Heatmap memberikan gambaran visual tentang hubungan antara kategori dan metrik yang diamati.
Dari heatmap, kita dapat melihat dengan jelas bagaimana nilai-nilai metrik (jumlah data, jumlah nilai unik, dan frekuensi nilai teratas) berbeda untuk setiap kategori.
